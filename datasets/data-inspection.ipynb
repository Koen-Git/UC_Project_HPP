```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
DATASET_NAMES = ['fc', 'kc', 'poa', 'sp']
DATASETS = {}
for ds in DATASET_NAMES:
    DATASETS[ds] = np.load(f'{ds}/data.npz')
    print("Dataset:", ds)
    print(DATASETS[ds]['X_train'].shape)
    if ds == 'kc':
        for i in range(DATASETS[ds]['X_train'].shape[1]):
            print("Col: ", i)
            print("Min:", np.min(DATASETS[ds]['X_train'][:,i]))
            print("Max:", np.max(DATASETS[ds]['X_train'][:,i]))
            # Print number of unique values in column i
            print("Unique values:", len(np.unique(DATASETS[ds]['X_train'][:,i])), "\n")
```
